### Define a function for displaying images

* We will want to look at images as we examine and analyze the data, so here will will define a function for displaying images. 
* This will handle getting the stretch of the image correct. 
* Feel free to play with the settings and colormap once you get the hang of the process.



```python
def imdisplay(image, v1perc=10, v2perc=95):
    '''
    display an image 
    OPTIONAL KEYWORD PARAMETERS
    v1perc: one end of the colormap assigned to the v1perc percent lowest flux 
    v2perc: the other end of the colormap assigned to the v2perc percent highest flux    
    '''
    # make sure image is an np array
    nimage = np.array(image)
    # determine the pixel values at the 10th and 95th percentile
    v1 = scoreatpercentile(nimage,v1perc)
    v2 = scoreatpercentile(nimage,v2perc)
    # display using imshow
    #
    # you can play with alternate cmaps in the function below, such as "viridis" or "gray"
    # The 'gray_r' color map reverses the color-scale so that dark display pixels are the brightest in the image
    #
    # vmin and vmax set the min and max pixel values that
    # will be mapped to the extremes of the colormap
    plt.imshow(nimage,cmap='gray_r',vmin=v1,vmax=v2)
    plt.colorbar(fraction=.08)
```

## Derivation

Write equation for gain $g$ using its units. Represent the mean of detected electrons $e^-$ as $N_{e^-}$ and mean ADU in an image \(ADU\) as \(N_{ADU}\). Call this equation (1).
\begin{equation} 
    g\frac{e^-}{ADU} \Rightarrow g = \frac{N_{e^-}}{N_{ADU}}
\end{equation}
Write an equation to represent the variance using a rearranged (1). Variance is standard deviation squared, or $\sigma^2$. Subscript $e$ represents the mean of photons and subscript $ADU$ represents the mean of the ADU (LaTeX doesn't like double subscripts so I am shortening $N_{e^-}$ and $N_ADU$ to something it processes more easily here). I then rearrange the equation so I can use it for substitution more easily. Call this equation (2).
\begin{equation}
    g^2{\sigma_N}^2 = {\sigma_{ADU}}^2 \Rightarrow {\sigma_N}^2 = \frac{{\sigma_{ADU}}^2}{g^2}
\end{equation}
Recall that the variance is equal to the mean in Poisson statistics. I can express this as ${\sigma_N}^2 = N_{e^-}$, substitute into (2), and equate that to a rearranged (1).
$$N_{e^-} = \frac{{\sigma_{ADU}}^2}{g^2} \Rightarrow \frac{N_{ADU}}{g} = \frac{{\sigma_{ADU}}^2}{g^2}$$
Simplification gives me a derivation I can use to find the gain with the information I have.
$$g = \frac{{\sigma_{ADU}}^2}{N_{ADU}}$$

```python
#*********** 
# http://labrigger.com/blog/2010/07/30/measuring-the-gain-of-your-imaging-system/

# your code goes here
# gain = how many electrons per ADU
# Ne = NADU*g (NADU = mean counts)
#poisson operates on photons- unc. on Ne = sqrt(Ne) 

num_adu = np.mean(data_flatb) # get the mean ADU
sigma_adu = np.std(data_flatb) # get the standard deviation of the ADU

temp = (sigma_adu)**2 # square the standard deviation to make it the variance
gain = temp/num_adu # divide variance by mean (following derivation from cell above!)

gain_actual = header_flatb['EGAIN'] # get the actual gain from the header

diff_vals = abs(gain - gain_actual) # absolute value of the difference between calculated and actual gain
avg_vals = (gain + gain_actual)/2 # get the average of the two values
diff_percent = 100*(diff_vals/avg_vals) # calculate percentage difference

print("gain = ", gain )
print("gain from header keyword = ", gain_actual)
print("percentage difference = ", diff_percent)
```
